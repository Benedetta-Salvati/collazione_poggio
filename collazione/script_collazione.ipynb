### 2.1. Normalizzazione

Rimuoviamo i segni di interpunzione, regolarizziamo le minuscole e maiuscole, e gli spazi.

```python
import glob, re, os

path = '../data/poggio1/'  # porre la directory come variabile

os.makedirs(path + 'norm', exist_ok=True)  # creare una nuova cartella

files = [os.path.basename(x) for x in glob.glob(path+'*.txt')]  # selezionare tutti i files della directory precedente

for file in files:  # per tutti i files della directory, fino alla fine
    
    ### leggere i files
    file_opened = open(path+file, 'r', encoding='utf-8') # aprire i files in modalità lettura 'r' (read)
    content = file_opened.read()  # leggere il contenuto dei files
    
    ### mettere tutto in minuscolo
    lowerContent = content.lower() 
    
    ### rimuovere tutto ciò che non è un carattere o uno spazio
    noPunct_lowerContent = re.sub(r'[^\w\s]','',lowerContent) 
    
    ### regolarizzare gli spazi/rimuovere spazi superflui
    regularSpaces_noPunct_lowerContent = " ".join(noPunct_lowerContent.split())
    
    ### creare un nuovo file
    filename = file.split('.')[0]
    new_file = open(path+'norm/' + filename + '_norm.txt', 'w', encoding='utf-8') # open the new file in mode 'w' (write)
    
    ### riscrivere il contenuto nel nuovo file
    new_file.write(regularSpaces_noPunct_lowerContent) 
    
    ### chiudere il file
    new_file.close()
    
print('Fatto ? Bene !')
```
